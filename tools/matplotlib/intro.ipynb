```python
%matplotlib inline 
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
# first method
plt.plot(x, y)
```

```python
# second method

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y)
```

```python
#  third method
fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
x = np.linspace(0, 10, 100  
)
y = np.cos(x)
fig, ax = plt.subplots(figsize=(10, 10)) # figsize=(width, height)
ax.plot(x, y)
ax.set_title('Title')
ax.set(
    xlabel='x-axis',
    ylabel='y-axis',
    xlim=(0, 10),
    ylim=(-2, 2),
    xticks=range(0, 11, 2),
    yticks=np.linspace(-2, 2, 9),
    aspect=0.5  
)

# plt.show()
fig.savefig('assets/sample-plot.png')
```
