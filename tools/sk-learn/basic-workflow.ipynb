## 1. Getting the data ready

```python
import pandas as pd
import numpy as np

heart_disease = pd.read_csv("assets/csv/heart-disease.csv")
heart_disease
```

```python
# X is the data (features) and y is the target (labels)

#  we will first train the model on the data and then test it on the data
# features will be used to predict the target while the target will contain the final required output
X = heart_disease.drop("target", axis=1)
y = heart_disease["target"]
```

 ## 2. Choosing the right model

<a href="https://scikit-learn.org/stable/tutorial/machine_learning_map/index.html" target="_blank">Scikit-Learn machine learning map</a>

<img src="./assets/images/sklean-map.png" style="background:white" width=1000/>

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100)

clf.get_params()
# we will use the RandomForestClassifier algorithm to train the model on the data
```

## 3. Fit the model/algorithm and use it to make predictions on our data


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
clf.fit(X_train, y_train)
```

```python
X_train
```

```python
clf.predict(np.array([1, 2, 3, 4]))
```

```python
y_preds = clf.predict(X_test)
y_preds
```

```python
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

## 4. Evaluating a model

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score


print(classification_report(y_test, y_preds))
```

```python
accuracy_score(y_test, y_preds)
```

```python
confusion_matrix(y_test, y_preds)
```

## 5. Improve a model

```python
np.random.seed(42)
for i in range(10, 100, 10):
    print(f"Trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f"Model accuracy on test set: {clf.score(X_test, y_test) * 100:.2f}%")
    print("")
```

## 6. Save and load a trained model

```python
import pickle as pkl

model_path = "./assets/models/heart-disease-model.pkl"
pkl.dump(clf, open(model_path, "wb"))
```

```python

# Load the model
loaded_model = pkl.load(open(model_path, "rb"))
loaded_model.score(X_test, y_test)
```
