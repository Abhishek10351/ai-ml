```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
heart_disease = pd.read_csv("./assets/heart-disease.csv")
heart_disease.head()
```

```python
heart_disease["age"].plot.hist(bins=10)
```

```python
heart_disease.plot.hist(figsize=(10, 30), subplots=True)
```

### Which one to use ? (pyplot or matplotlib OO)
* To plot something quickly, use pyplot method
* To plot something more advanced, use the OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]

over_50.plot(kind="scatter", x="age", y="chol", c="target", cmap="cool")
over_50.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", cmap="cool", ax=ax)
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))
```
