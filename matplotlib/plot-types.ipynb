### Add different types of plots in matplotlib (mainly using numpy arrays)

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1000, 100)
y = np.square(x)
```

### Different types of plots:

```python
# Line plot
fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
 # Scatter plot
fig, ax = plt.subplots()
ax.scatter(x, y)
```

```python
# Bar plot
fig, ax = plt.subplots()
prices = {'apple': 1.99, 'banana': 0.99, 'orange': 1.49}
ax.bar(prices.keys(), prices.values())
ax.set(ylabel='Price', title='Fruit prices', ylim=[0, 2])
```

```python
# Horizontal bar plot
fig, ax = plt.subplots()
ax.barh(list(prices.keys()), list(prices.values()))
ax.set(xlabel='Price', title='Fruit prices', xlim=[0, 2])
```

```python
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x, bins=40)
```

### Subplots 

```python
# Subplots

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 10))
x1 = np.linspace(0, 1, 100)
x2 = np.random.randn(100)

ax1.plot(x1, np.square(x1))
ax1.scatter(np.random.random(30), np.random.random(30), color='purple')
ax2.scatter(np.random.random(30), np.random.random(30))
ax3.bar(prices.keys(), prices.values())
ax4.hist(x2, bins=40)
```

```python
# Subplots option 2
fig, ax = plt.subplots(2, 2, figsize=(10, 10))
x1 = np.linspace(0, 1000, 100)
x2 = np.random.randn(100)

ax[0, 0].plot(x1, np.square(x1))
ax[0, 1].scatter(np.random.random(30), np.random.random(30))
ax[1, 0].bar(prices.keys(), prices.values())
ax[1, 1].hist(x2, bins=40)
```
