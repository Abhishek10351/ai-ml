### Plotting from pandas DataFrame

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
car_data = pd.read_csv("assets/car-sales.csv")
car_data
```

```python
day_period = 30
ts = pd.Series(np.exp(np.linspace(0,5,num=day_period)), index=pd.date_range('1/1/2024', periods=day_period))
ts.plot()
ts.head()
```

```python
day_period = 100
ts = pd.Series(np.random.randn(day_period), index=pd.date_range('1/1/2024', periods=day_period))
ts = ts.cumsum()
ts.plot()
```

### PLot using Pandas DataFrame

```python
car_data
```

```python
car_data["Price"]=car_data['Price'].str.replace("[$,.]", '', regex=True).str[:-2].astype(int)
car_data
```

```python
car_data["Sale Date"] = pd.date_range("1/1/2024", periods=len(car_data))

car_data["Total Sales"] = car_data["Price"].cumsum()
car_data
```

```python
car_data.plot(x="Sale Date", y="Total Sales", kind="line")
```

```python
car_data.plot(x="Odometer (KM)", y="Price", kind="scatter")
```
