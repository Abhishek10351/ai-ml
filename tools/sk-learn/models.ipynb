```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

%matplotlib inline
```

## 2.1 Picking a machine learning model for a regression problem

```python
# California Housing Dataset
from sklearn.datasets import fetch_california_housing

housing = fetch_california_housing()
housing
```

```python
housing_df = pd.DataFrame(housing.data, columns=housing.feature_names)
housing_df
```

```python
housing_df["Target"] = pd.Series(housing.target)
housing_df
```

```python
from sklearn.linear_model import Ridge

np.random.seed(42)

X = housing_df.drop("Target", axis=1)
y = housing_df["Target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = Ridge()
model.fit(X_train, y_train)

model.score(X_test, y_test)  # coefficient of determination R^2 of the prediction
```

Trying out different models, to get better results according to our needs

We try an ensemble method, a RandomForestClassifier, and a Support Vector Machine (SVM) for the heart disease dataset.
```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()

np.random.seed(42)

X = housing_df.drop("Target", axis=1)
y = housing_df["Target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model.fit(X_train, y_train)

model.score(X_test, y_test)
```

## 2.2 Picking a machine learning model for a classification problem

```python
from sklearn.datasets import load_iris

iris = load_iris()
iris
```

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df
```

```python
iris_df["Target"] = pd.Series(iris.target)
iris_df
```

```python
from sklearn.svm import SVC

np.random.seed(42)

X = iris_df.drop("Target", axis=1)
y = iris_df["Target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = SVC()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.naive_bayes import GaussianNB

np.random.seed(42)

X = iris_df.drop("Target", axis=1)
y = iris_df["Target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = GaussianNB()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
