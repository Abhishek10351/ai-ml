### Which one to use ? (pyplot or matplotlib OO)
* To plot something quickly, use pyplot method
* To plot something more advanced, use the OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]

over_50.plot(kind="scatter", x="age", y="chol", c="target", cmap="cool")
over_50.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", cmap="cool", ax=ax)

```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="cool")
# add the colorbar for cmap

plt.colorbar(scatter)

ax.set(
    title="Heart Disease and Cholesterol Levels",
    xlabel="Age",
    ylabel="Cholesterol"
)
# Hide Legend
ax.legend(*scatter.legend_elements(), title="Target").set_visible(False)
ax.axhline(over_50["chol"].mean(), linestyle="--", color="salmon")
```

```python
# plot with cholesterol levels and thalach
fig , (ax0,ax1) = plt.subplots(2, figsize=(10,10), sharex=True)
# Heart Disease and Cholestrol Levels
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"],c=over_50["target"], cmap="cool")
plt.colorbar(scatter)
ax0.set(title="Cholestrol Levels", ylabel="Cholestrol")
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")
ax0.legend(*scatter.legend_elements(), title="Target")
# ax0.set_xlim([50,75])
ax0.set_ylim([150,350])

# Heart Disease and Max Heart Rates
scatter = ax1.scatter(x=over_50["age"],y= over_50["thalach"], c=over_50["target"], cmap="winter")
plt.colorbar(scatter)
ax1.set(title="Thalach Levels/ Max Heart Rate",xlabel="Age", ylabel="Thalach")
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--")
ax1.legend(*scatter.legend_elements(), title="Target")

fig.suptitle("Heart Disease Analysis", fontsize="20", fontweight="bold", color="red"
)
fig.savefig("assets/chol and thalach.png")
```
